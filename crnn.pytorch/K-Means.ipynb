'K-means'

```python
from sklearn.cluster import KMeans

import pandas as pd

import numpy as np

from sklearn.decomposition import PCA

import matplotlib.pyplot as plt

#import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = np.random.random((300,4))
pd_data = pd.DataFrame(data)
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(data)
kmeans.labels_
```

```python
pca = PCA(2).fit(data)
reduced = pca.transform(data).T
```

```python
reduced.T[0]
```

```python
list(kmeans.labels_)
```

```python
plt.title("Clustering - user driven decision making", fontsize = 14)
plt.scatter(reduced[0], reduced[1], c=kmeans.labels_, cmap=plt.cm.Set1)
plt.legend(bbox_to_anchor = (1.05, 1))
plt.xlabel('Highest Variance direction', fontsize = 14) ; plt.ylabel('Second Highest Variance', fontsize = 14)
#plt.legend( bbox_to_anchor=(1.05,1))
```
