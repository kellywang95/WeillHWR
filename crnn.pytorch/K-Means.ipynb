```python
f, ax = plt.subplots(1,1, figsize=(40,40), )


data = np.random.random((200,2))
pd_data = pd.DataFrame(data)
kmeans = KMeans(n_clusters=3, random_state=0).fit(data)
kmeans.labels_
reduced = pd_data
s = [400 for n in range(len(reduced[0]))]
ax.scatter(reduced[0], reduced[1], c=kmeans.labels_, marker="o", s=s)
ax.set_title("Machine Learning Assisted Decision Making", fontsize = 100)
ax.legend(bbox_to_anchor = (1.05, 1))
ax.set_xlabel('Population by features', fontsize = 70)
ax.set_ylabel('Fatality Rate', fontsize = 70)
ax.set_xticklabels(["Sector_" + str(k) for k in range(8)], {'fontsize': 50})

plt.savefig("possible_cluster_image.png")
    
```
